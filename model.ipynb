```python
# =========================
# 1️⃣ Install (if needed)
# =========================
# !pip install matplotlib pillow

# =========================
# 2️⃣ Imports
# =========================
import os
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

# =========================
# 3️⃣ Dataset path
# =========================
path = '/content/data'   # your extracted folder

# =========================
# 4️⃣ Get all valid images
# =========================
images = [f for f in os.listdir(path) if f.lower().endswith(('.jpg', '.jpeg', '.png'))]

if len(images) == 0:
    raise Exception("❌ No images found!")

print("✅ Total images found:", len(images))

# =========================
# 5️⃣ Show random images
# =========================
fig, ax = plt.subplots(1, 5, figsize=(20, 5))
fig.suptitle("Sample Images", fontsize=18)

for i in range(5):
    img_path = os.path.join(path, images[np.random.randint(0, len(images))])

    try:
        img = np.array(Image.open(img_path))
        ax[i].imshow(img)
        ax[i].axis('off')
    except:
        print("⚠️ Skipping:", img_path)

plt.show()

# =========================
# 6️⃣ Convert images to dataset (for ML)
# =========================
IMG_SIZE = 128

X = []

for img_name in images:
    img_path = os.path.join(path, img_name)

    try:
        img = Image.open(img_path).convert('RGB')
        img = img.resize((IMG_SIZE, IMG_SIZE))
        img = np.array(img)
        X.append(img)
    except:
        continue

X = np.array(X)

print("✅ Dataset shape:", X.shape)

# =========================
# 7️⃣ Normalize
# =========================
X = X / 255.0

# =========================
# 8️⃣ Dummy labels (since no classes)
# =========================
y = np.zeros(len(X))  # all same class

# =========================
# 9️⃣ Build simple CNN model
# =========================
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense

model = Sequential([
    Conv2D(32, (3,3), activation='relu', input_shape=(128,128,3)),
    MaxPooling2D(2,2),

    Conv2D(64, (3,3), activation='relu'),
    MaxPooling2D(2,2),

    Flatten(),
    Dense(128, activation='relu'),
    Dense(1, activation='sigmoid')   # single class
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

model.summary()

# =========================
# 🔟 Train model
# =========================
model.fit(X, y, epochs=3, batch_size=32)

print("✅ Training complete!")
```
